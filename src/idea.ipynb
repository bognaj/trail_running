# Theory

## Energy supply

To describe the power supply (aerobic and anaerobic) we will use the hydraulic analogy. 
* we assume that aerobic energy is of infinite capacity and flows at a maximal rate $\sigma$. $\sigma$ is the energetic equivalent per unit time of VO2, the volume of oxygen used by unit of time. We can assume that 1l of oxyegn produces 20kJ [1], 20.9 kJ [2]. 
* anaerobic energy has finite capacity modeled by a container of height 1 and surface $A_p$

**Accumulated oxygen deficit**

When the anaerobic container starts depleting by a height $h$ the accumulated oxygen deficit is:
$$e_{an}^{0} - e_{an} = A_p h$$
where $e_{an}^{0} = A_p$ is the initial supply of anaerobic energy. 

**Improvements of basic hydarulic analogy from [1]**
* initial phase: some time at the beginning is needed to reach $VO2_{max}$
* drop in VO2 at the end of the race, when $e_{an}$ falls below $e_{crit}$ 
* recreation of anaeribic energy when slowing down - to be considered. However it should be rather the drop in vertical displacement than the horizontal one.

* check Minetti's formulas for available fraction of $VO2_{max}$ based on race length and altitude [2]

**Energetic cost of walking and running [2], [3]**

Cost of walking/running ($C_w$/$C_r$) is calculated as the ratio of rate of metabolic energy expenditure to the nominal speed. 
Based on the Minetti's findings [2] we can express them as the function of gradient $i$ ($5^{th}$ order polynomials). In [3] the energetics costs has been directly measured for the greater slopes ([2] up to 24.2 degrees, [3] up to 45 degrees). 

What can be interesting is switching to walking when the gradient is steep. As shown in [3], considering the same vertical speed, walking is energetically cheaper from running.
